```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 示例数据（您需要替换为您的实际数据）
a = torch.tensor([1.2, 2.3, 3.1, 4.5, 5.0])
b = torch.tensor([2.1, 2.9, 3.7, 4.8, 5.6])
c = torch.tensor([1.5, 2.7, 3.4, 4.2, 4.9])

# 将数据放入一个列表中
data = [a, b, c]
labels = ['a', 'b', 'c']
```

```python
# 计算每列的平均值和标准误差（SE）
means = [torch.mean(col).item() for col in data]
std_errors = [torch.std(col).item() / np.sqrt(len(col)) for col in data]

# 计算95%置信区间的误差条
confidence_intervals = [1.96 * se for se in std_errors]
```

```python

# 绘制图形
plt.figure(figsize=(8, 6))
plt.bar(labels, means, yerr=confidence_intervals, capsize=5, alpha=0.6, color='skyblue')
plt.xlabel('Data Columns')
plt.ylabel('Mean Value')
plt.title('Error Bar Plot with 95% Confidence Interval')
plt.grid(True)

# 显示图形
plt.show()
```
